```python


import pandas as pd
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from imblearn.combine import SMOTEENN
import shap
import numpy as np
import matplotlib.pyplot as plt
```

```python
df0=pd.read_csv("employee-attrition.csv")
print(df0.info())
```

```python
df=df0.copy()
le=LabelEncoder()

df[["Over18","OverTime","MaritalStatus","JobRole","Gender","Department","Attrition","BusinessTravel","EducationField"]]=df[["Over18","OverTime","MaritalStatus","JobRole","Gender","Department","Attrition","BusinessTravel","EducationField"]].apply(lambda x:le.fit_transform(x))
```

```python
columns=df.columns
for col in columns:
    count=len(df[col].unique())
    if count<=10:
        print(col,count)
```

```python
df.drop(["StandardHours","EmployeeCount","EmployeeNumber"],axis=1,inplace=True)
```

```python
#Nominal categorical columns => Attrition, BusinessTravel,Department,EducationField,Gender,JobRole,MaritalStatus,Over18,OverTime
```

```python
#model = XGBClassifier(use_label_encoder=False, eval_metric='logloss',scale_pos_weight = 1233/237)

model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

x=df.drop(["Attrition"],axis=1)
y=df["Attrition"]
smote_enn = SMOTEENN()
xresampled, yresampled = smote_enn.fit_resample(x, y)
xtrain,xtest,ytrain,ytest=train_test_split(xresampled,yresampled,test_size=0.2,random_state=42)
model.fit(xtrain, ytrain)

ypred = model.predict(xtest)
print(classification_report(ytest, ypred))
```

```python
df.groupby("Attrition").size()
```

```python
xgb.plot_importance(model)
```

```python
explainer = shap.Explainer(model, xtrain)
shap_values = explainer(xtest)
shap.summary_plot(shap_values, xtest,plot_type="bar")
```

```python
#Order of features which on increasing, decrease attrition the most
#High Monthly Income, Hourly rate and Monthly rate, employees have attrition less >
#Employees with higher stock option level have attribution less >
#Older employees and employees with more experience in the company tend to get attrition less , suggesting that these employees stand the test of time by the company and are profitable >
#Lots of years since promotion increase risk of attrition >
```

```python
mean_shap_values = np.mean(shap_values.values, axis=0)  # shape: (n_features,)

# Put into DataFrame
shap_df = pd.DataFrame({
    'feature': x.columns,
    'mean_shap_value': mean_shap_values,
    'abs_mean_shap_value': np.abs(mean_shap_values)
})

# Sort by absolute value (for ranking)
shap_df = shap_df.sort_values(by='abs_mean_shap_value', ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
colors = ['green' if val > 0 else 'red' for val in shap_df['mean_shap_value']]

plt.barh(shap_df['feature'], shap_df['mean_shap_value'], color=colors)
plt.xlabel('Average SHAP Value per Feature')
plt.title('Average SHAP Value per Feature (Positive = Class 1, Negative = Class 0)')
plt.gca().invert_yaxis()  # highest values on top
plt.axvline(x=0, color='black', linestyle='--')
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
