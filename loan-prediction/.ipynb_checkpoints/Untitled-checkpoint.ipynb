```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("train.csv",header=0)

print(df.drop(["Loan_ID"],axis=1,inplace=True))
```

```python
df.head()
```

```python
def trim_mean(a):
    b=a.sort_values(ascending=True)
    l=len(a)
    return b.iloc[int(0.1*l):int(0.9*l)]
```

```python
df.info()
def give_chi_sqval(column1,column2):
    k=min(len(column1.unique()),
         len(column2.unique()))
    return chi2_contingency(pd.crosstab(column1,column2))[0]/(len(column1)*(k-1))**0.5
```

```python
for column in df.select_dtypes(include=["object"]):
    print(column,df[column].unique())
```

```python
df_dropped=df.dropna()
def cut_bin(column,num_bins):
    minn=column.min()
    maxx=column.max()
    step=(maxx-minn)/num_bins
    bins=[0+(i*step) for i in range(num_bins)]+[maxx]
    labels=["Low","High","Medium"]
    #print(column.isnull().sum(),"yoyoyo")
    return pd.cut(column,bins=bins,labels=labels)

for column in df.select_dtypes(exclude=["object"]):
    if column!="Credit_History":
        df[column+"_bins"]=cut_bin(df[column],3)
df.info()
```

```python
category_columns=df.select_dtypes(include=["object","category"])
#print(category_columns)
chi2_table=pd.DataFrame(0.0,index=category_columns.columns,columns=category_columns.columns)
for column1_name in category_columns:
    column1=df[column1_name]
    for column2_name in category_columns:
        column2=df[column2_name]
        value=give_chi_sqval(column1,column2)
        chi2_table.loc[column1_name,column2_name]=value


print(chi2_table)
```

```python
for column in category_columns:
    most_dependent_column=chi2_table[column].nlargest(2).index[1]
    print(column+"("+str(df[column].isnull().sum())+")"," --> ",most_dependent_column+"("+str(df[most_dependent_column].isnull().sum())+") "+str(chi2_table[column][most_dependent_column]))
    
```

```python
#gender,dependents,self_employed
#df.drop("CoapplicantIncome_bins",axis=1,inplace=True)
df.dropna(subset=["Married"],inplace=True)

df["Gender_na"]=df.groupby("Married")["Gender"].transform(lambda a:a.mode().iloc[0])
df["Dependents_na"]=df.groupby("Married")["Dependents"].transform(lambda a:a.mode().iloc[0])
#df["Self_Employed"]=df.groupby("CoapplicantIncome_bins")["Self_Employed"].transform(lambda a:a.mode().iloc[0])
df["Self_Employed_na"]=df.groupby("Dependents")["Self_Employed"].transform(lambda a:a.mode().iloc[0])

df["LoanAmount_bins_na"]=df.groupby("ApplicantIncome_bins")["LoanAmount_bins"].transform(lambda a:a.mode().iloc[0])
df["CoapplicantIncome_bins_na"]=df.groupby("LoanAmount_bins")["CoapplicantIncome_bins"].transform(lambda a:a.mode().iloc[0])

df["Loan_Amount_Term_bins_na"]=df.groupby("Dependents")["Loan_Amount_Term_bins"].transform(lambda a:a.mode().iloc[0])
#df["Credit_History_bins"].fillna(df["Credit_History_bins"].mode().iloc[0],inplace=True)


df["Gender"]=df["Gender"].fillna(df["Gender_na"])
df["Dependents"]=df["Dependents"].fillna(df["Dependents_na"])
df["Self_Employed"]=df["Self_Employed"].fillna(df["Self_Employed_na"])
df["CoapplicantIncome_bins"]=df["CoapplicantIncome_bins"].fillna(df["CoapplicantIncome_bins_na"])
df["LoanAmount_bins"]=df["LoanAmount_bins"].fillna(df["LoanAmount_bins_na"])
df["Loan_Amount_Term_bins"]=df["Loan_Amount_Term_bins"].fillna(df["Loan_Amount_Term_bins_na"])
```

```python

pre=df.groupby("CoapplicantIncome_bins")["CoapplicantIncome"].transform(lambda a:trim_mean)
df["CoapplicantIncome"].fillna(pre,inplace=True)

pre=df.groupby("LoanAmount_bins")["LoanAmount"].transform(trim_mean)
df["LoanAmount"].fillna(pre,inplace=True)

pre=df.groupby("Loan_Amount_Term_bins")["Loan_Amount_Term"].transform(trim_mean)
df["Loan_Amount_Term"].fillna(pre,inplace=True)

df["Credit_History"].fillna(df["Credit_History"].mode().iloc[0],inplace=True)
```

```python
df.info()
df.dropna(inplace=True)
```

```python
#Categorical -> Gender, Married, Self_Employed,Dependents,Education,Property_Area
#Numerical -> ApplicantIncome, CoapplicantIncome,LoanAmount, Loan_Amount_Term, Credit_History
"""
le=LabelEncoder()
df_original=df.copy()
#df[["Gender","Married","Self_Employed","Dependents","Education"]]=le.fit_transform(df[["Gender","Married","Self_Employed","Dependents","Education"]])
categorical=["Gender","Married","Self_Employed","Dependents","Education"]
for category in categorical:
    df[category]=le.fit_transform(df[category])
"""
le=LabelEncoder()
col_to_encode=["Gender","Married","Self_Employed","Dependents","Education","Property_Area","Loan_Status"]
for col in col_to_encode:
    df[col]=le.fit_transform(df[col])

columns=list(df.columns)
for col in columns:
    print(col)
    if "_bins" in col or "_na" in col:
        df.drop(col,axis=1,inplace=True)
```

```python
X_categorical_columns=['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area',"Credit_History"]
X_columns=['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area',"ApplicantIncome","CoapplicantIncome",
                      "LoanAmount","Loan_Amount_Term","Credit_History"]
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X=df[X_columns]
X_original=X.copy()
X=pd.get_dummies(X,columns=X_categorical_columns,drop_first=True)
for X_column in X.columns:
    X[X_column]=X[X_column].astype(float)
Y=df["Loan_Status"]
Y=Y.astype(int)
Y_original=Y.copy()
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
scaler = StandardScaler()
smote = SMOTE(random_state=42)
X_train_sm, Y_train_sm = smote.fit_resample(X_train, Y_train)
X_train_sm_scaled = scaler.fit_transform(X_train_sm)
X_test_scaled=scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_sm_scaled, Y_train_sm)
```
